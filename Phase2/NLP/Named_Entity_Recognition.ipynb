```python
# First NER Program Using spaCy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in 2002."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
nlp = spacy.load("en_core_web_sm")

text = "Microsoft acquired Activision Blizzard for $68.7 billion."

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text,
        ent.label_,
        spacy.explain(ent.label_)
    )
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Elon Musk met Sundar Pichai and Satya Nadella.
"""

doc = nlp(text)

persons = []

for ent in doc.ents:
    if ent.label_ == "PERSON":
        persons.append(ent.text)

print(persons)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Jeff Bezos founded Amazon in Seattle in 1994."

doc = nlp(text)

entities = []

for ent in doc.ents:
    entities.append({
        "text": ent.text,
        "label": ent.label_
    })

print(entities)
```

```python
# Practical: NER Visualization
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

text = """
Elon Musk founded SpaceX in California.
"""

doc = nlp(text)

displacy.serve(doc, style="ent")
```

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    aggregation_strategy="simple"
)

text = """
Jeff Bezos founded Amazon in Seattle.
"""

result = ner(text)

for item in result:
    print(item)
```

```python
# Resume Parser (Real Industry Project)

import spacy

nlp = spacy.load("en_core_web_sm")

resume = """
Anuruddh Kumar worked at Microsoft.
He completed B.Tech from AKTU.
"""

doc = nlp(resume)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
