```python
# Practical 2: Count Number of Tokens

print("Token length : ", len(tokens))
```

```python
for i, el in enumerate(tokens):
    print(i, el)
```

```python
# Practical 3: Create Token IDs

vocab = {}

for i, word in enumerate(sorted(set(tokens))):
    vocab[word] = i

print(vocab)
```

```python
text = "cat dog cat bird"

tokens = text.split()

vocab = {
    "cat":0,
    "dog":1,
    "bird":2
}

encoded = [vocab[token] for token in tokens]

print(encoded)
```

```python
# Practical 5: Character Tokenization
text = "NLP"
tokens = list(text)
print(tokens)
```

```python
# Practical 6: Character encoding
text = "NLP"

tokens = list(text)

vocab = {char:i for i,char in enumerate(tokens)}

encoded = [vocab[t] for t in tokens]

print(encoded)
```

```python
# Practical 8 : Sentence Tokenize
from nltk.tokenize import sent_tokenize


text = """
I love NLP
Tokenization is important.
Let's learn it.
"""

sentences = sent_tokenize(text)

print(sentences)
```

```python
# Practical 9: Remove Stopwords

from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
```

```python
text = "I am learning Natural Language Processing"

tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))

filtered = [word for word in tokens if word.lower() not in stop_words]

print(filtered)
```

```python
# 1. Hugging Face BERT Tokenizer (WordPiece)

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = "ChatGPT is amazing"

tokens = tokenizer.tokenize(text)

print(tokens)
```

```python
# Token IDs from BERT

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = "I love NLP"

ids = tokenizer.encode(text)

print(ids)
```

```python
# 5. Decode IDs Back to Text

decoded = tokenizer.decode(ids)

print(decoded)
```
