```python
from nltk.tokenize import word_tokenize

text = "The machine learning model accurately predicted customer churn."

tokens = word_tokenize(text)

tags = nltk.pos_tag(tokens)

print(tags)
```

```python
text = "Apple reported record revenue."

tokens = nltk.word_tokenize(text)

tags = nltk.pos_tag(tokens)

for word, tag in tags:
    print(f"{word:15} {tag}")
```

```python
# Find the noun from the tokens
text = "The model accurately predicted the result."

tokens = nltk.word_tokenize(text)

tags = nltk.pos_tag(tokens)

noun = []

for word, tag in tags:
    if tag.startswith("NN"):
        noun.append(word)

print(noun)
```

```python
# Practical: Count POS Tags

import nltk
from collections import Counter

text = "The quick brown fox jumps over the lazy dog"

tokens = nltk.word_tokenize(text)

tags = nltk.pos_tag(tokens)

tag_list = [tag for word, tag in tags]

counter = Counter(tag_list)

print(counter)
```

```python
# Practical: spaCy POS Tagging (Industry Standard)

import spacy

nlp = spacy.load("en_core_web_sm")

text = "Apple Inc. reported record quarterly revenue."

doc = nlp(text)

for token in doc:
    print(
        token.text,
        token.pos_,
        token.tag_
    )
```
